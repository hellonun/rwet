```python
import random
import textwrap

subjects = """Rest of You Light and Interactivity Sound Studio Reading and Writing Electronic Text Choreographic Interventions"""
professorsfirst = """Dan Tom Luke Allison Mimi """
professorslast = """O'Sullivan Igoe Dubois Parrish Yin"""
descriptions = """investigates techniques in sound art in acoustic, analog electronic, and digital domains

re-conceives interactive media as a form of choreographic intervention

 how lighting is used for utilitarian, expressive, and informational purposes

introduces the Python programming language as a tool for reading and writing digital text

nature of these illusions has long been described in mystical practices but is now increasing corroborated by modern research such as neuroscience, behavioral economics, social psychology, embodied cognition, and evolutionary psychology. 
"""

subjectswords = subjects.split()
professorsfirstwords = professorsfirst.split()
professorslastwords = professorslast.split()
descriptionswords = descriptions.split()



```

```python
import random
stanza_count = 10
for i in range(stanza_count):
    credit = random.randrange(1,5)
    print()
    subject = (random.choice(subjectswords) + " " + random.choice(subjectswords) + " " + random.choice(subjectswords))
    units = " (" + str(credit) + ".0 units)"
    print(subject.title() + units )
    print(random.choice(professorsfirstwords) + " " + random.choice(professorslastwords) + "\n")
    random.shuffle(descriptionswords)
    description = ("This course " + textwrap.fill(" ".join(descriptionswords), 40)) 
    description = '. '.join(i.capitalize() for i in description.split('. '))
    print (description [0:300] + "... [read more]"+ "\n"+ "\n"+ "\n"+ "\n"+ "\n")
```
